```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
temp_values = list(range(1, 21))
per_values = [round(0.0001 * i, 4) for i in range(0,60,2)]
```

```python
detection_error = []
x = 1
for y in per_values:
    # print(y)
    filename = './metrics/met_{}_{}_.txt'.format(x, y)
    print(filename)
    if not os.path.exists(filename):
        continue
    with open(filename, 'r') as file:
        content = file.readline()
        error = float(content.split(':')[1])
        print(error)
        detection_error.append(error)
detection_error = np.array(detection_error)
```

```python
print(detection_error)
```

```python
fig = plt.figure(1)
plt.plot(per_values,detection_error)
plt.xlabel('perturbation')
plt.ylabel('detection error')
plt.title('Detection Error vs perturbation')
plt.show()
```
